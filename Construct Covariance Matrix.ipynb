```python
for i in np.arange(1, df3.shape[0]):
    for j in np.arange(1, df3.shape[1]):
        if math.isnan(df3.iloc[i, j]):
            cpos = [j]
            c =list(df3.columns[cpos])
            rpos = [i]
            r=list(df3.iloc[rpos,0])
            df3.iloc[i, j]=mean.loc[r, c].iloc[0,0]
```
