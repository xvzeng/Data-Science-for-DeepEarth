```
import io
pixel = pd.read_excel(io.BytesIO(uploaded['pixel.xlsx']))
```

```
pixel.head()
```

```
# Create an 721 * 361 coordinate matrix
# With latitude from -380 to 380 and longitude from -180 to 180
# Pre-set the original concentration for each cell as 0
location = pd.DataFrame(np.zeros((721, 361)))
```

```
location.head()
```

```
location.shape
```

```
# Rename the columns based on integers of longitude.
location.columns=np.arange(-180, 181, 1)
```

```
# Rename the rows based on integers of latitude.
location.index=np.arange(-360, 361, 1)
```

```
location.head()
```

```
# Round the coordiantes in pixel file to integers and emove NAs
pixel_round=pixel.round({'LATITUDE': 0, 'LONGITUDE': 0}).dropna()
# Convert the round results to integer
pixel_round.LATITUDE=pixel_round.LATITUDE.astype(int)
pixel_round.LONGITUDE=pixel_round.LONGITUDE.astype(int)
```

```
pixel_round.head()
```

```
# Get the average values of element concentration based on the same coordinates.
ave=pixel_round.groupby(['LATITUDE','LONGITUDE'], as_index=False).agg({'Mg#':'mean'})
```

```
ave.head()
```

```
ave.shape
```

```
ave.isna().sum().sum()
```

```
print(ave.iloc[0,1])
print(ave.iloc[1,1])
print(ave.iloc[2,1])
print(ave.iloc[3,1])
#ave.iloc[k, 1] == b
```

```
ave.iloc[1,2]
```

```
# Fill in the values to coordinate matrix based on the concentration mean from pixel data
for a in np.arange(-360, 361, 1):
  for b in np.arange(-180, 181, 1):
    for k in np.arange(0, 638, 1):
      if ave.iloc[k, 0] == a and ave.iloc[k, 1] == b:
        location.iloc[a, b]=ave.iloc[k,2]
```

```
from itertools import chain

def draw_map(m, scale=0.2):
    # draw a shaded-relief image
    m.shadedrelief(scale=scale)
    
    # lats and longs are returned as a dictionary
    lats = m.drawparallels(np.linspace(-90, 90, 13))
    lons = m.drawmeridians(np.linspace(-180, 180, 13))

    # keys contain the plt.Line2D instances
    lat_lines = chain(*(tup[1][0] for tup in lats.items()))
    lon_lines = chain(*(tup[1][0] for tup in lons.items()))
    all_lines = chain(lat_lines, lon_lines)
    
    # cycle through these lines and set the desired style
    for line in all_lines:
        line.set(linestyle='-', alpha=0.3, color='w')
```

```
fig = plt.figure(figsize=(15, 12), edgecolor='w')
m = Basemap(projection='cyl', resolution=None,
            llcrnrlat=-90, urcrnrlat=90,
            llcrnrlon=-180, urcrnrlon=180, )
draw_map(m)
```

```
# Extract the 
lat = ave['LATITUDE'].values
lon = ave['LONGITUDE'].values
concentration = ave['Mg#'].values
#area = cities['area_total_km2'].values
```

```
# 1. Draw the map background
fig = plt.figure(figsize=(15, 15))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=50, lon_0=-100,
            width=8E6, height=8E6)
m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# 2. scatter element data, with color reflecting concentration
# and size reflecting area
m.scatter(lon, lat, latlon=True,
          c=concentration,s=concentration,
          cmap='Reds', alpha=0.5)

# 3. create colorbar and legend
plt.colorbar(label=r'$concentration$')
plt.clim(30, 110)

# make legend with dummy points
for a in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.8, s=a,
                label=str(a) + ' km$^2$')
plt.legend(scatterpoints=1, frameon=False,
           labelspacing=1, loc='lower left');
```
